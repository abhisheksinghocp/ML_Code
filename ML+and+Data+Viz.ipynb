fig = plt.figure(1, figsize=(4, 3))

```python

pca = decomposition.PCA(n_components=3)
pca.fit(X)
X = pca.transform(X)
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```
