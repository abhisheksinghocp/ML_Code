```
import pandas as pd
data_cluster = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00396/Sales_Transactions_Dataset_Weekly.csv')
fillter_data_x_y_data = data_cluster[data_cluster.columns[0:53]]
fillter_data_x_data = data_cluster[data_cluster.columns[1:53]]
fillter_data_y_target = data_cluster[data_cluster.columns[0]]
```

```
import random
random.seed(1234)
```

```
# clustering dataset
# determine k using elbow method
 
from sklearn.cluster import KMeans
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score

# k means determine k
distortions = []
K = range(2,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(fillter_data_x_data)
    kmeanModel.fit(fillter_data_x_data)
    distortions.append(
        sum(np.min(cdist(fillter_data_x_data, kmeanModel.cluster_centers_, 'euclidean'), 
                   axis=1)) / fillter_data_x_data.shape[0])
    label = kmeanModel.labels_
    sil_coeff = silhouette_score(fillter_data_x_data, label, metric='euclidean')
    print("For k={}, The Silhouette Coefficient is {}".format(k, sil_coeff))


 
# Plot the elbow
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```
# Initializing KMeans
kmeans = KMeans(n_clusters=5)
# Fitting with inputs
kmeans = kmeans.fit(fillter_data_x_data)
# Predicting the clusters
labels = kmeans.predict(fillter_data_x_data)
# Getting the cluster centers
C = kmeans.cluster_centers_
#print(labels)
```
